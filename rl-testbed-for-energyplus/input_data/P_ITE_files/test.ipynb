```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib widget

# Read the CSV file
df_list = []
filenames = []
for i in range(6):
    filename = 'IT_Load_' + str(i) + '.csv'
    filenames.append(filename)
    df_list.append(pd.read_csv(filename))

# Plot the data in separate figures
for i, df in enumerate(df_list):
    plt.figure()
    plt.plot(df.index, df['Load']) 
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.title(f'Plot of data from {filenames[i]}')
    plt.show()
```
